```python
import re
import string
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

#tahap pemotongan string berdasarkan tiap kata yang Menyusun sebuah kalimat
def tokenizing(data):

    data = re.sub(r'[^a-zA-Z0-9\s]','', data)
    data = re.sub(r'\d+', '', data)
    data = data.translate(str.maketrans('', '', string.punctuation))
    line = data.split('\n')
    lines = np.asarray(line)
    lines = np.char.strip(lines)
    lines = lines[lines != '']

    
    return lines
```

```python
text = open('orange.txt', 'r')
corpus = tokenizing(text.read().lower())

print(corpus)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
response = vectorizer.fit_transform(corpus)
```
